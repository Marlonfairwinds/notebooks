```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Function to get ticket audits
def get_ticket_audit(ticket_id):
    url = f'https://{subdomain}.zendesk.com/api/v2/tickets/{ticket_id}/audits'
    response = requests.get(url, auth=auth)
    if response.status_code == 200:
        return response.json()
    else:
        return None


ticket_descriptions = []
# testing_ticket = ["29834","29866","29823","29866"]

for ticket_id in ticket_ids[1100:]:
    ticket = get_ticket_audit(ticket_id)
    if ticket:
        audits = ticket["audits"]
        ticket_description = ''  # Initialize ticket description for each ticket
        for audit in audits:
            events = audit["events"]
            for event in events:
                body_text = event.get("body", "")
                testing1 = BeautifulSoup(body_text, "html.parser").get_text().strip()

                if testing1:
                    ticket_description += testing1 + '\n\n'  # Append ticket description
        ticket_descriptions.append(ticket_description)

# Create a DataFrame with each ticket's information in a separate cell
audit_information_df = pd.DataFrame(ticket_descriptions, columns=["Ticket Description"])
```

```python
ticket_structure_df = pd.concat([ticket_information_df, audit_information_df], axis=1)
```
