```python
import boto3
from datetime import datetime, timedelta

# Create a Boto3 client for CloudWatch
client = boto3.client('cloudwatch', region_name=region)

# Set the start time and end time for the metric data retrieval
start_time = datetime.utcnow() - timedelta(minutes=30)
end_time = datetime.utcnow()

# Define the instance ID for which you want to retrieve metrics


# Define the filters for the metric data
filters = [
    {
        'Name': 'myd4cc',
        'Value': instanceID
    }
]

# Define the historical metrics to retrieve
historical_metrics = [
    {
        'Id': 'cpu_utilization',
        'MetricStat': {
            'Metric': {
                'Namespace': 'AWS/EC2',
                'MetricName': 'CPUUtilization',
                'Dimensions': filters
            },
            'Period': 300,
            'Stat': 'Average',
            'Unit': 'Percent'
        }
    }
]

# Call the GetMetricData API to retrieve the metric data
response = client.get_metric_data(
    aws_access_key_id=access_key_id,
    aws_secret_access_key=secret_access_key,
    region_name=region,
    StartTime=start_time,
    EndTime=end_time,
    MetricDataQueries=historical_metrics,
    ScanBy='TimestampAscending'
)

# Print the response
print(response)
```
